```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import mne
from mne.datasets import sample
from mne import make_fixed_length_epochs
from mne.time_frequency import psd_array_welch
from mne.decoding import (
    CSP,
    GeneralizingEstimator,
    LinearModel,
    Scaler,
    SlidingEstimator,
    Vectorizer,
    cross_val_multiscore,
    get_coef,
)
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
```

```python
dir = 'ds004504'

#num_sub = 'sub-001/eeg/sub-001_task-eyesclosed_eeg.set'
#data_path = f'{dir}/{num_sub}'

data_paths = []
eeg = '/eeg/'
task = '_task-eyesclosed_eeg.set'
num = ''

with open('participant_list.txt','r') as file:
    for line in file:
        num = line.rstrip()
        data_paths.append(f'{dir}/{num}{eeg}{num}{task}')


```

```python
low, high = 2,20
duration = 4 
overlap = 0.5 

all_epochs = []

for data_path in data_paths:
    raw = mne.io.read_raw_eeglab(data_path, preload=True)
    raw = raw.filter(low,high)
    epochs = make_fixed_length_epochs(raw, duration=duration, overlap=overlap, preload=True)
    data = epochs.get_data()
    all_epochs.append(epochs)
    
combined_epochs = mne.concatenate_epochs(all_epochs)
X = combined_epochs.get_data()

num_epochs = len(X)
y = np.zeros(num_epochs, dtype=int)

y[:36] = 0
y[36:65] = 1
y[65:] = 2
```

```python
def extract_features(X):
    psd_features = []
    for epoch in X:
        psd, freqs = psd_array_welch(epoch, sfreq=500, n_fft=2000)
        psd = psd.mean(axis=0)
        psd_features.append(psd)
    return np.array(psd_features)
    
X_features = extract_features(X)
X_train, X_test, y_train, y_test = train_test_split(X_features, y, test_size = 0.3, random_state = 42)

clf = make_pipeline(StandardScaler(), SVC(kernel='linear', probability=True))
clf.fit(X_train, y_train)

scores = cross_val_score(clf, X_features, y, cv=5)
print(f"Cross-validation accuracy: {np.mean(scores):.2f} ± {np.std(scores):.2f}")

test_score = clf.score(X_test, y_test)
print(f"Test set accuracy: {test_score:.2f}")

```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_features)
scores = cross_val_score(clf, X_pca, y, cv=5)
print(f"Cross-validation accuracy: {scores.mean():.2f} ± {scores.std():.2f}")
```
